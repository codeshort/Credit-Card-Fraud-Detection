```python
#import some necessay packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the datset
data = pd.read_csv('creditcard.csv')
```

```python
#exploring the dataset
print(data.columns)
```

```python
print(data.shape)
```

```python
print(data.describe())
```

```python
data = data.sample(frac=0.1,random_state = 27)
print(data.shape)
```

```python
data.hist(figsize= (20,20))
plt.show()
```

```python
#Determine number of fraud cases
Frauds = data[data['Class']==1]
Valid = data[data['Class']==0]

fraud_to_valid = len(Frauds)/float(len(Valid))

print("Fraud Caes: {}".format(len(Frauds)))
print("Valid Cases: {}".format(len(Valid)))
print(fraud_to_valid)
```

```python
correlation_matrix = data.corr()
fig = plt.figure(figsize = (12,9))
sns.heatmap(correlation_matrix, vmax = .8, square =True)
#plt.show()
```
